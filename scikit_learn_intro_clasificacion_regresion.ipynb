```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
 
 
```

Creamos un conjunto de datos ficticios de clientes
Las columnas son: 'Frecuencia de compra (dÃ­as)', 'Monto total gastado (USD)', 'NÃºmero de visitas'
Y la clase serÃ¡ la categorÃ­a: 'Comprador frecuente', 'Comprador ocasional', 'Nuevo cliente'

```python
data = {
    "Frecuencia_compra": [
        30,
        90,
        5,
        120,
        30,
        10,
        60,
        200,
        45,
        15,
        10,
        50,
        250,
        100,
        200,
    ],
    "Monto_gastado": [
        500,
        100,
        150,
        50,
        200,
        450,
        350,
        20,
        120,
        400,
        300,
        250,
        10,
        100,
        15,
    ],
    "Visitas": [10, 3, 20, 1, 12, 8, 5, 1, 7, 12, 10, 6, 0, 4, 2],
    "Clase": [
        "Frecuente",
        "Ocasional",
        "Nuevo",
        "Ocasional",
        "Frecuente",
        "Nuevo",
        "Ocasional",
        "Nuevo",
        "Frecuente",
        "Frecuente",
        "Nuevo",
        "Frecuente",
        "Nuevo",
        "Ocasional",
        "Nuevo",
    ],
}

df = pd.DataFrame(data)
print(df.head())
 
```

```python
# Variables predictoras (X) y la variable objetivo (y)
x = df[["Frecuencia_compra", "Monto_gastado", "Visitas"]]  # Caracteristicas
y = df["Clase"]  # Variable objetivo etiquetas
# Dividir el conjunto de datos en entrenamiento y prueba
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42
)  # 80% entrenamiento, 20% prueba
# Crear el modelo KNN
knn = KNeighborsClassifier(n_neighbors=3)  # Usar 3 vecinos
# Entrenar el modelo
knn.fit(x_train, y_train)  # Entrenamos el modelo con los datos de entrenamiento.
# Hacer predicciones
y_pred = knn.predict(x_test)  # Usamos el modelo para predecir las etiquetas del conjunto de prueba.
# Evaluar el modelo
accuracy = accuracy_score(y_test, y_pred)  # Comparamos las predicciones del modelo con las etiquetas reales.
print(f"Accuracy: {accuracy * 100:.2f}%")  # Imprimir la precisiÃ³n
# Crear la matriz de confusiÃ³n
# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
# Visualizar la matriz de confusiÃ³n
plt.figure(figsize=(8, 6))
sns.heatmap(
    cm,
    annot=True,
    fmt="d",
    cmap="Blues",
    xticklabels=knn.classes_,
    yticklabels=knn.classes_,
)
plt.title("Matriz de ConfusiÃ³n")
plt.xlabel("PredicciÃ³n")
plt.ylabel("Realidad")
plt.show()
# Visualizar la distribuciÃ³n de las caracterÃ­sticas
plt.figure(figsize=(12, 8))
sns.histplot(
    df,
    x="Frecuencia_compra",
    hue="Clase",
    multiple="stack",
    kde=True,
    bins=30,
)
plt.title("DistribuciÃ³n de Frecuencia de Compra")
plt.xlabel("Frecuencia de Compra")
plt.ylabel("Frecuencia")
plt.legend(title="Clase")
plt.show()
# Visualizar la distribuciÃ³n de las caracterÃ­sticas
plt.figure(figsize=(12, 8))
sns.histplot(
    df,
    x="Monto_gastado",
    hue="Clase",
    multiple="stack",
    kde=True,
    bins=30,
)
plt.title("DistribuciÃ³n de Monto Gastado")
plt.xlabel("Monto Gastado")
plt.ylabel("Frecuencia")
plt.legend(title="Clase")
plt.show()
# Visualizar la distribuciÃ³n de las caracterÃ­sticas
plt.figure(figsize=(12, 8))
sns.histplot(
    df,
    x="Visitas",
    hue="Clase",
    multiple="stack",
    kde=True,
    bins=30,
)
plt.title("DistribuciÃ³n de Visitas")
plt.xlabel("Visitas")
plt.ylabel("Frecuencia")
plt.legend(title="Clase")
plt.show()
# Visualizar la distribuciÃ³n de las caracterÃ­sticas
plt.figure(figsize=(12, 8))
sns.histplot(
    df,
    x="Frecuencia_compra",
    hue="Clase",
    multiple="stack",
    kde=True,
    bins=30,
)
plt.title("DistribuciÃ³n de Frecuencia de Compra")
plt.xlabel("Frecuencia de Compra")
plt.ylabel("Frecuencia")
plt.legend(title="Clase")
plt.show()
# Visualizar la distribuciÃ³n de las caracterÃ­sticas
plt.figure(figsize=(12, 8))
sns.histplot(
    df,
    x="Monto_gastado",
    hue="Clase",
    multiple="stack",
    kde=True,
    bins=30,
)
plt.title("DistribuciÃ³n de Monto Gastado")
plt.xlabel("Monto Gastado")
plt.ylabel("Frecuencia")
plt.legend(title="Clase")
plt.show()
# Visualizar la distribuciÃ³n de las caracterÃ­sticas
plt.figure(figsize=(12, 8))
sns.histplot(
    df,
    x="Visitas",
    hue="Clase",
    multiple="stack",
    kde=True,
    bins=30,
)
plt.title("DistribuciÃ³n de Visitas")
plt.xlabel("Visitas")
plt.ylabel("Frecuencia")
plt.legend(title="Clase")
plt.show()
# Visualizar la distribuciÃ³n de las caracterÃ­sticas
plt.figure(figsize=(12, 8))
sns.histplot(
    df,
    x="Frecuencia_compra",
    hue="Clase",
    multiple="stack",
    kde=True,
    bins=30,
)
plt.title("DistribuciÃ³n de Frecuencia de Compra")
plt.xlabel("Frecuencia de Compra")
plt.ylabel("Frecuencia")
plt.legend(title="Clase")
plt.show()
```
